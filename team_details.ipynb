```python
gcolumns=['Name','Reactions','GKDiving','GKHandling','GKKicking','GKPositioning','GKReflexes','Position']
gkdata=data[gcolumns]
```

```python
gkdata=gkdata[gkdata.GKKicking>=70]
gkdata.head()
```

```python
gkdata.shape
```

```python
gkdata['Summation']=gkdata['GKDiving']+gkdata['GKHandling']+gkdata['GKPositioning']+gkdata['GKReflexes']
```

```python
gkdata.shape
```

```python
gkdata.head()
```

```python
gkdatasub=gkdata[gkdata.Position=='SUB']
gkdata=gkdata[gkdata.Position!='SUB']
```

```python
gkdata=gkdata.sort_values(['GKKicking','Summation','Reactions'],ascending=(False,False,False))
gkdatasub=gkdatasub.sort_values(['GKKicking','Summation','Reactions'],ascending=(False,False,False))
```
